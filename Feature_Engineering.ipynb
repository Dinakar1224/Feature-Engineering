**1. What is a parameter?**

A parameter is a value that defines a characteristic of a model or function. In machine learning, parameters are learned from data during training (for example, weights in linear regression).

**2. What is correlation?**

Correlation measures the strength and direction of a relationship between two variables. It ranges from -1 (perfect negative correlation) to +1 (perfect positive correlation).

**2b. What does negative correlation mean?**

A negative correlation means as one variable increases, the other tends to decrease (inverse relationship).

**3. Define Machine Learning. What are the main components in Machine Learning?**

Machine Learning (ML) is a field of AI where computers learn patterns from data without being explicitly programmed.

Main components:

Data - the input for learning.

Model - the structure used to learn (for example, decision tree, neural network).

Loss function - measures model error.

Optimizer/algorithm - updates model parameters to reduce loss.

**5. What are continuous and categorical variables?**

Continuous variable: Numeric value with infinite possibilities (for example, height, weight).

Categorical variable: Represents categories or groups (for example, gender, color).

**6. How do we handle categorical variables in Machine Learning? What are common techniques?**

Common techniques:

Label Encoding - assign numbers to categories (for example, red=0, green=1)

One-Hot Encoding - create binary columns for each category

**7. What do you mean by training and testing a dataset?**

Training dataset: Data used to teach the model.

Testing dataset: Data used to evaluate model performance.

**8. What is sklearn.preprocessing?**

A Scikit-learn module that provides functions for data transformation like scaling, encoding, normalization, etc.

**9. What is a test set?**

A test set is the portion of data not seen by the model during training. It is used to measure how well the model generalizes to new data.

7. Tune hyperparameters.

**15. What is causation? Explain difference between correlation and causation with an example.**

Causation: One variable directly affects another.

Correlation: Variables move together but not necessarily cause-effect.

Example: Ice cream sales and drowning cases may increase together (correlation) because of summer, but ice cream sales do not cause drowning (no causation).

**16. What is an Optimizer? What are different types of optimizers? Explain each with an example.**

An optimizer updates model parameters to minimize loss.

Types:

SGD (Stochastic Gradient Descent) - updates with small data batches.

Adam - adaptive learning rate, combines RMSProp and momentum.

RMSProp - adjusts learning rate for each parameter.

Example:

from tensorflow.keras.optimizers import Adam
optimizer = Adam(learning_rate=0.001)

Feature scaling transforms data to a similar scale (for example, 0-1).
It helps improve model convergence and prevents features with large values from dominating.

**25. Explain data encoding**

Data encoding is the process of converting categorical variables into numerical form so machine learning models can process them.
Common types of encoding:

Label Encoding: Each category is assigned a unique integer.

One-Hot Encoding: Creates a new binary column for each category (1 if present, 0 otherwise).

Ordinal Encoding: Assigns integers to categories where the order matters (e.g., low=1, medium=2, high=3).
